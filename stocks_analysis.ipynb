```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbols for the stocks
tickers = ["AAPL", "GOOGL", "MSFT", "AMZN", "TSLA"]

# Retrieve the historical stock prices for the past 5 years
data_original = yf.download(tickers, period="5y")

# Print the data
print(data_original)
```

```python
data_original
```

```python
# Create a figure with a wider size
plt.figure(figsize=(20, 6))
# Create a line plot of the "open" prices
plt.plot(data_original["Open"]['AAPL'])

# Create a line plot of the "open" prices for each stock
plt.plot(data_original["Open"]["AAPL"], label="AAPL")
plt.plot(data_original["Open"]["GOOGL"], label="GOOGL")
plt.plot(data_original["Open"]["MSFT"], label="MSFT")
plt.plot(data_original["Open"]["TSLA"], label="TSLA")


# Add a title and labels to the plot
plt.title("Open Prices")
plt.xlabel("Date")
plt.ylabel("Price")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

```python

# Select the data for a specific time range
start_date = "2019-10-01"
end_date = "2022-12-31"
data = data_original.loc[start_date:end_date]

# Create a figure with a wider size
plt.figure(figsize=(20, 6))

# Create a line plot of the "open" prices for each stock
plt.plot(data["Open"]["AAPL"], label="AAPL")
plt.plot(data["Open"]["GOOGL"], label="GOOGL")
plt.plot(data["Open"]["MSFT"], label="MSFT")

# Add a title and labels to the plot
plt.title("Open Prices (2021)")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()
# Show the plot
plt.show()
```
